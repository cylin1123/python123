### AppleNews Crawler

```python
import requests
from bs4 import BeautifulSoup
```

```python
#res = requests.get('https://tw.news.appledaily.com/politics/realtime/20181020/1451284/')
res = requests.get('https://tw.news.appledaily.com/international/realtime/20181020/1451280/')
#res.text
```

```python
soup = BeautifulSoup(res.text)
soup
```

```python
title = soup.find('h1').text
print(title)
crateAt = soup.find(name='div',attrs={"class":"ndArticle_creat"}).text
print(crateAt)
```

```python
contentArea = soup.find(name = 'div',attrs={"class":"ndArticle_margin"})
contentArea
```

```python
content = contentArea.find('p').text
print(content)
```

```python
content = contentArea.find('p')
for tag in content.select('a'):
    tag.decompose()
print(content.text)
```

```python
for tag in content.select('strong'):
    tag.decompose()

print(content.text)
```
